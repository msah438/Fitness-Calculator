```python
from tkinter import *
class fit:
    def __init__(self):
        window = Tk()
        window.geometry("500x500")
        window.title("Fitness Calculator")
        self.f = Frame(window , height = 500,width  = 500 , bg ="pink2")

        self.f.pack()
        self.lb1 =Label(self.f , text = "Name",font=("Helvetica 10 bold"),bg = 'white' , fg ='black').place (x = 10 , y = 10)
        self.lb2 =Label(self.f , text = "Age",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 250 , y = 10)
        self.e1 = Entry(self.f)
        self.e1.place(x=60,y=10)
        self.e2 = Entry(self.f)
        self.e2.place(x=300,y=10)
        self.lb3 =Label(self.f , text = "Gender",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 50)
        var1 = StringVar()
        
        self.R1 = Radiobutton(self.f , text = 'Male' ,font=("Helvetica 10 bold"),variable = var1 , value  ="Male")
        self.R1.place(x=80 , y =50)
        self.R2 = Radiobutton(self.f , text = 'Female' ,font=("Helvetica 10 bold"),variable = var1 , value  ="Female")
        self.R2.place(x=150 , y =50)
        
        self.lb4 =Label(self.f , text = "Weight :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 100)
        self.lb5 =Label(self.f , text = "Height :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 130)
        self.lb6 =Label(self.f , text = "BP low :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 160)
        self.lb7 =Label(self.f , text = "BP high :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 190)
        self.lb8 =Label(self.f , text = "Pulse Rate :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 220)
        self.lb1 =Label(self.f , text = "RBC Count :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 250)
        self.lb1 =Label(self.f , text = "WBC Count :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 280)
        self.lb1 =Label(self.f , text = "Platelets :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 310)
        self.lb1 =Label(self.f , text = "HB :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 340)
        self.lb1 =Label(self.f , text = "URIC Acid :",font=("Helvetica 10 bold"),bg = 'white' , fg ='black').place (x = 10 , y = 370)
        self.lb1 =Label(self.f , text = "Cholestrol :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 10 , y = 400)
        
        
        #-------------UNITS------------------------
        
        self.lb20 =Label(self.f , text = "KG",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=100)
        self.lb20 =Label(self.f , text = "METERS",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=130)
        self.lb20 =Label(self.f , text = "cells/mL",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=250)
        self.lb20 =Label(self.f , text = "cells/mL",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=280)
        self.lb20 =Label(self.f , text = "cells/mL",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=310)
        self.lb20 =Label(self.f , text = "g/dL",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=340)
        self.lb20 =Label(self.f , text = "mg/dL",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=370)
        self.lb20 =Label(self.f , text = "mg/dL",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=240,y=400)
        
        self.e3 = Entry(self.f)
        self.e3.place(x=110,y=100)
        self.e4 = Entry(self.f)
        self.e4.place(x=110,y=130)
        self.e5 = Entry(self.f)
        self.e5.place(x=110,y=160)
        self.e6 = Entry(self.f)
        self.e6.place(x=110,y=190)
        self.e7 = Entry(self.f)
        self.e7.place(x=110,y=220)
        self.e8 = Entry(self.f)
        self.e8.place(x=110,y=250)
        self.e9 = Entry(self.f)
        self.e9.place(x=110,y=280)
        self.e10 = Entry(self.f)
        self.e10.place(x=110,y=310)
        self.e11 = Entry(self.f)
        self.e11.place(x=110,y=340)
        self.e12 = Entry(self.f)
        self.e12.place(x=110,y=370)
        self.e13 = Entry(self.f)
        self.e13.place(x=110,y=400)
        self.b1 = Button(self.f , text = "Generate Report" , font=("Helvetica 11 bold"),bg = 'lightblue' , fg ='black',command = self.report).place(x = 350 , y=450)
        self.b2 = Button(self.f , text ="VIEW" , font=("Helvetica 11 bold"),bg = 'lightblue' , fg ='black',command = self.view).place(x = 10 , y=450)
        window.mainloop()
        
    def report(self):
        root = Tk()
        root.geometry("500x500")
        root.title("Report")
        self.f2 =Frame (root , height =500 ,width = 500 ,bg='silver')
        self.f2.pack()
        
        
        #-------------------------NAME & AGE---------------------------
        name = self.e1.get()
        age = self.e2.get()
        self.lbname = Label(self.f2 , text = name,font=("Helvetica 10 bold"), bg = 'white' , fg ='black')
        self.lbname.place(x=80,y=20)
        self.lbage = Label(self.f2 , text = age,font=("Helvetica 10 bold"), bg = 'white' , fg ='black')
        self.lbage.place(x=300,y=20)
        
        
        
        
        self.lb11 =Label(self.f2 , text = "NAME :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 20 , y = 20)
        self.lb12 =Label(self.f2 , text = "AGE :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 240 , y = 20)
        
        self.lb1 =Label(self.f2 , text = "BMI (Body Mass Index) :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 20 , y = 70)
        self.lb2 =Label(self.f2 , text = "BP (High/Medium/Low) :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 20 , y = 100)
        self.lb3 =Label(self.f2 , text = "Pulse Rate (High/Medium/Low) :", font=("Helvetica 10 bold"),bg = 'white' , fg ='black').place (x = 20 , y = 130)
        self.lb4 =Label(self.f2 , text = "RBC Count (High/Medium/Low) :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 20 , y = 160)
        self.lb5 =Label(self.f2 , text = "WBC (High/Medium/Low) :", font=("Helvetica 10 bold"),bg = 'white' , fg ='black').place (x = 20 , y = 190)
        self.lb6 =Label(self.f2 , text = "Platelets (High/Medium/Low) :", font=("Helvetica 10 bold"),bg = 'white' , fg ='black').place (x = 20 , y = 220)
        self.lb7 =Label(self.f2 , text = "HB (High/Medium/Low) :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 20 , y = 250)
        self.lb8 =Label(self.f2 , text = "URIC Acid (High/Medium/Low) :",font=("Helvetica 10 bold"), bg = 'white' , fg ='black').place (x = 20 , y = 280)
        self.lb10 =Label(self.f2 , text = "Cholestrol (High/Medium/Low) :", font=("Helvetica 10 bold"),bg = 'white' , fg ='black').place (x = 20 , y = 310)
        
        button = Button(self.f2, text="NEW REPORT",font=("Helvetica 10 bold"), bg = 'white', fg="red",command=self.__init__)
        button.place(x=380 , y=400)   
        
        button = Button(self.f2, text="SAVE",font=("Helvetica 10 bold"), bg = 'white', fg="red",command=self.save)
        button.place(x=320 , y=400) 
        
        
        #-------------UNITS------------------------
        
        self.lb20 =Label(self.f2 , text = "KG/m^2",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=380,y=70)
        self.lb21 =Label(self.f2 , text = "YEARS",font=("Helvetica 10 bold"), bg = 'lightblue' , fg ='black').place (x=330,y=20)
        
        
        
        
        self.e11 = Entry(self.f2)
        self.e11.place(x=250,y=70)
        self.e22 = Entry(self.f2)
        self.e22.place(x=250,y=100)
        self.e33 = Entry(self.f2)
        self.e33.place(x=250,y=130)
        self.e44 = Entry(self.f2)
        self.e44.place(x=250,y=160)
        self.e55 = Entry(self.f2)
        self.e55.place(x=250,y=190)
        self.e66 = Entry(self.f2)
        self.e66.place(x=250,y=220)
        self.e77 = Entry(self.f2)
        self.e77.place(x=250,y=250)
        self.e88 = Entry(self.f2)
        self.e88.place(x=250,y=280)
        self.e99 = Entry(self.f2)
        self.e99.place(x=250,y=310)
        
        
        #-----------------BMI CALCULATION------------------------
        height = float(self.e4.get())
        weight = float(self.e3.get())
        bmindex = weight / (height)**2
        self.e11.insert(END,bmindex)
        
        
        #-----------------BP CALCULATION-------------------------
        bpLow = int(self.e5.get())
        bpHigh = int(self.e6.get())
        if((bpLow>=40  and bpLow <= 60) and (bpHigh >=70 and bpHigh<=90)):
            self.e22.insert(END,'Low')
        if((bpLow>=60  and bpLow <= 80) and (bpHigh >=90 and bpHigh<=120)):
            self.e22.insert(END,'Normal')
        if((bpLow>=80  and bpLow <= 90) and (bpHigh >=120 and bpHigh<=140)):
            self.e22.insert(END,'Pre Hypertension')
        
        
        #-------------------Pulse Rate ---------------------------
        
        pulseRate = float(self.e7.get())
        
        if (pulseRate>=95 and pulseRate<=133):
            self.e33.insert(END , 'Light')
        if (pulseRate>=133 and pulseRate<=152):
            self.e33.insert(END , 'Moderate')
        if (pulseRate>=152 and pulseRate<=190):
            self.e33.insert(END , 'Hard')
            
        #-------------------RBC------------
        
        Rbc = float(self.e8.get())
        if(Rbc < 4500):
            self.e44.insert(END , 'Low')
        elif(Rbc>=4500 and Rbc< 5500):
            self.e44.insert(END,'Medium')
        else:
            self.e44.insert(END , 'High')
            
        #--------------------WBC----------------------------------

        Wbc = float(self.e9.get())
        if(Wbc < 5000):
            self.e55.insert(END , 'Low')
        elif(Wbc>=5000 and Wbc<= 10000):
            self.e55.insert(END,'Medium')
        else:
            self.e55.insert(END , 'High')
       
      #--------------------PLATELETS----------------------------------
           
        Platelets = float(self.e10.get())
        if(Platelets < 200):
            self.e66.insert(END , 'Low')
        elif(Platelets>=200 and Platelets<= 500):
            self.e66.insert(END,'Medium')
        else:
            self.e66.insert(END , 'High')
            
    #--------------------HB----------------------------------
            
            
        Hb = float(self.e11.get())
        if(Hb < 14):
            self.e77.insert(END , 'Low')
        elif(Hb>=14 and Hb<=17):
            self.e77.insert(END,'Medium')
        else:
            self.e77.insert(END , 'High')
            
           
           
       #--------------------URIC ACID----------------------------------
    
        UricAcid = float(self.e12.get())
        if(UricAcid < 0.18):
            self.e88.insert(END , 'Low')
        elif(UricAcid>=0.18and UricAcid<=0.54):
            self.e88.insert(END,'Medium')
        else:
            self.e88.insert(END , 'High')
    
    
    #--------------------CHOLESTROL----------------------------------
    
        Cholestrol = float(self.e13.get())
        if(Cholestrol <= 1.7):
            self.e99.insert(END , 'Low')
        elif(Cholestrol>1.7 and Cholestrol<=2.1):
            self.e99.insert(END,'Medium')
        else:
            self.e99.insert(END , 'High')
                
        
        root.mainloop()
        
    def save(self):
        root2 = Tk()
        root2.geometry("500x500")
        root2.title("Report")
        self.f3 =Frame (root2 , height =500 ,width = 500 ,bg='silver')
        self.f3.pack()
        self.esave = Entry(self.f3)
        self.esave.place(x=250 , y=50)
        self.lblsave = Label(self.f3,text="ENTER FILE NAME: ",font=("Helvetica 10 bold"), bg = 'white', fg="red").place(x=100,y=50)
        
        
        self.btn2 = Button(self.f3,text="SAVE",font=("Helvetica 10 bold"), bg = 'white', fg="red",command = self.data)
        self.btn2.place(x=350,y=450)
        root2.mainloop()
        
    def data(self):
        
        #-----------------BMI CALCULATION------------------------
        height = float(self.e4.get())
        weight = float(self.e3.get())
        bmindex = weight / (height)**2
        name = self.e1.get()
        age = self.e2.get()
        bpLow = int(self.e5.get())
        bpHigh = int(self.e6.get())
        pulseRate = float(self.e7.get())
        Rbc = float(self.e8.get())
        Wbc = float(self.e9.get())
        Platelets = float(self.e10.get())
        Hb = float(self.e11.get())
        UricAcid = float(self.e12.get())
        Cholestrol = float(self.e13.get())
        fname= self.esave.get()
        
        with open(fname + '.txt','w')as wr:
            
            wr.write('NAME : '+ str(name) + '\n')
            wr.write('AGE : '+ str(age) + '\n')
            wr.write('BMI : '+ str(bmindex) + '\n')
            wr.write('BP LOW : ' + str(bpLow) + '\n')
            wr.write('BP HIGH : ' + str(bpHigh) + '\n')
            wr.write('PULSE RATE : ' + str(pulseRate) + '\n')
            wr.write('RBC : ' + str(Rbc) + '\n')
            wr.write('WBC : ' + str(Wbc) + '\n')
            wr.write('PLATELETS : ' + str(Platelets) + '\n')
            wr.write('HB : ' + str(Hb) + '\n')
            wr.write('URIC ACID : ' + str(UricAcid) + '\n')
            wr.write('CHOLESTROL : ' + str(Cholestrol))
            
            
        Label(self.f3,text = 'File Saved Succesfully' , fg='red' , bg='silver' , font=("Helvetica 20 bold")  ).place(x=120, y=250)
    def view(self):
        root3 = Tk()
        root3.geometry("500x500")
        root3.title("View")
        self.f4 =Frame (root3 , height =500 ,width = 500 ,bg='silver')
        self.f4.pack()
        self.eview = Entry(self.f4)
        self.eview.place(x=250 , y=50)
        self.lblview = Label(self.f4,text="ENTER FILE NAME: ",font=("Helvetica 10 bold"), bg = 'white', fg="red").place(x=100,y=50)
        
        
        self.btn2 = Button(self.f4,text="VIEW",font=("Helvetica 10 bold"), bg = 'white', fg="red",command = self.display)
        self.btn2.place(x=350,y=450)
        root3.mainloop()
    def display(self):
        root4 = Tk()
        root4.geometry("500x500")
        root4.title("Display")
        self.f5 =Frame (root4 , height =500 ,width = 500 ,bg='silver')
        self.f5.pack()
        self.t1 = Text(self.f5 , height = 500 , width = 500 , bg ='lightblue')
        self.t1.grid(column =0, row= 0)
        fname2 = self.eview.get()
        with open(fname2 + '.txt','r')as fr:
            a=fr.read()
            self.t1.insert(END,a)
        root4.mainloop()    
object = fit()
```
